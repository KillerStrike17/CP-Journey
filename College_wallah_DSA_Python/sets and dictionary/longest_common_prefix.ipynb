```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        min_s = min(strs,key= len)
        print(min_s)
        # strs = sorted(strs, key = len)
        counter = 0
        while counter !=len(min_s):
            check_val = min_s[:counter+1]
            substring_flag = False
            for i in strs:
                if check_val != i[:counter+1]:
                    substring_flag = True
            if substring_flag :
                return min_s[:counter]
            
            counter +=1
        return min_s[:counter]

        
        
```
