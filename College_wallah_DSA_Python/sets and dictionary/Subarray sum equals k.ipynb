```python
class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        n = len(nums)
        results = []
        counter = 0
        for i in range(n):
            for j in range(i+1,n+1):
                if k == sum(nums[i:j]):
                    counter +=1
        return counter
        
```
