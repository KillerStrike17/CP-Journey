```python
class Solution:
    def findMaxAverage(self, nums: List[int], k: int) -> float:
        n = len(nums)
        curr_sum = 0
        for i in range(k):
            curr_sum +=nums[i]
        
        ans = curr_sum/k

        for i in range(k,n):
            curr_sum +=nums[i]
            curr_sum -=nums[i-k]
            if curr_sum/k > ans:
                ans = curr_sum/k
        return ans

        # if n ==1:
        #     return nums[0]
        # for i in range(n-k+1):
        #     value = sum(nums[i:i+k])/k
        #     if value > avg:
        #         avg = value
        # return avg

```
