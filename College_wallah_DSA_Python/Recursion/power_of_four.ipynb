```python
        class Solution:
    def isPowerOfFour(self, n: int) -> bool:
        if n ==0 or n==3 or n==2:
            return False
        if n==1:
            return True
        if n%4==1 or n%4==2 or n%4==3:
            return False
        return self.isPowerOfFour(n//4)
        
        
```
