```python
from typing import List

class Solution:
    def isValid(self, row, col, rows, cols):
        if row<=rows-1 and col<= cols-1 and row>=0 and col>=0:
            return True
        return False

    def spiralMatrixIII(self, rows: int, cols: int, rStart: int, cStart: int) -> List[List[int]]:
        
        matrix = []
        counter = 1
        for i in range(rows):
            row_matrix = []
            for j in range(cols):
                row_matrix.append(counter)
                counter +=1
            matrix.append(row_matrix)

        visit_counter = 0
        step_counter = 1
        visited_order = [[rStart,cStart]]
        
        while visit_counter != rows*cols-1:
            for i in range(cStart+1, cStart+step_counter+1):
                if self.isValid(rStart,i,rows,cols):
                    visit_counter+=1    
                    visited_order.append([rStart,i])
            if visit_counter == rows*cols:
                break
            print(visited_order)
            
            cStart += step_counter
            # rStart = rStart
            for i in range(rStart+1, rStart+step_counter+1):
                if self.isValid(i,cStart,rows,cols):
                    visit_counter+=1    
                    visited_order.append([i,cStart])
            if visit_counter == rows*cols:
                break
            print(visited_order)
            rStart+=step_counter
            step_counter+=1
            
            for i in range(cStart-1,cStart-step_counter-1, -1):
                if self.isValid(rStart,i,rows,cols):
                    visit_counter+=1    
                    visited_order.append([rStart,i])
            if visit_counter == rows*cols:
                break
            print(visited_order)
            
            cStart -=step_counter
            for i in range(rStart-1, rStart-step_counter-1,-1):
                if self.isValid(i,cStart,rows,cols):
                    visit_counter+=1    
                    visited_order.append([i,cStart])
            if visit_counter == rows*cols:
                break
            print(visited_order)
            rStart -=step_counter
            step_counter+=1
        return visited_order
            
                
            
```
