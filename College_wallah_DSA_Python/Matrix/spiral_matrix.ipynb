```python
class Solution:
    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:
        n = len(matrix)
        m = len(matrix[0])

        col_start, col_end = 0, m-1
        row_start, row_end = 0, n-1

        ans = []

        while len(ans)<m*n:
            # row start, col_start -> col_end
            for i in range(col_start,col_end+1):
                ans.append(matrix[row_start][i])


            if len(ans)==m*n:
                break
            
            # column end, row start -> row end
            for i in range(row_start+1,row_end+1):
                ans.append(matrix[i][col_end])
            
            if len(ans)==m*n:
                break

            # row end,col end-> col start
            for i in range(col_end-1,col_start-1,-1):
                ans.append(matrix[row_end][i])
            
            if len(ans)==m*n:
                break

            # column start, row end -> row start
            for i in range(row_end-1,row_start,-1):
                ans.append(matrix[i][col_start])
            
            row_start +=1
            col_start +=1 
            col_end -=1
            row_end -=1 

        return ans
```
