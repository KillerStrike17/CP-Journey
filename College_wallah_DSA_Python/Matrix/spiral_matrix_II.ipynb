```python
class Solution:
    def generateMatrix(self, n: int) -> List[List[int]]:

        matrix = []
        for i in range(n):
            row_matrix = []
            for j in range(n):
                row_matrix.append('*')
            matrix.append(row_matrix)
        # print(matrix)

        counter = 1
        row_start, row_end = 0,n-1
        column_start, column_end = 0, n-1
        
        while counter != n*n + 1:   
            for i in range(column_start, column_end+1):
                matrix[row_start][i]=counter
                counter +=1
            if counter ==n*n +1 :
                break

            for i in range(row_start+1, row_end+1):
                matrix[i][column_end]=counter
                counter +=1
            if counter ==n*n +1:
                break
            
            for i in range(column_end-1, column_start-1,-1):
                matrix[row_end][i]=counter
                counter +=1
            if counter ==n*n +1:
                break
            
            for i in range(row_end-1, row_start,-1):
                matrix[i][column_start]=counter
                counter +=1

            row_start += 1
            column_start += 1
            row_end -= 1
            column_end -=1
        return matrix
            




        output_matrix = []

        
```
