```python
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        i,counter,local_counter,n = 0,1,1,len(nums)
        for j in range(1,n):
            if nums[i]==nums[j] and local_counter<2:
                local_counter += 1
                counter +=1
                i+=1
                nums[i]=nums[j]
            elif nums[i]!=nums[j]:
                local_counter = 1
                i+=1
                nums[i]=nums[j]
                counter+=1
        return counter


        
```
