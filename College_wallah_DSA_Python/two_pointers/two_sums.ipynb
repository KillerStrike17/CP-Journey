```python
number = [5,7,11,20,24,30,45]
```

```python
def two_sum(number, target):
    n = len(number)
    i,j=0,n-1
    while i<j:
        curr_sum = number[i] + number[j]
        if curr_sum == target:
            return True
        elif curr_sum>target:
            j-=1
        else:
            i+=1
    return False
    
```

```python
two_sum(number, target=41)
```
