```python
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        i = 0
        j=1
        n = len(nums)
        counter= 1

        for j in range(1,n):
            if nums[i]!=nums[j]:
                i+=1
                nums[i]=nums[j]
                
                counter +=1

        return counter
```
