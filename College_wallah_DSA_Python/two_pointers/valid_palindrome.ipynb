```python

class Solution:
    def isPalindrome(self, s: str) -> bool:
        s = s.lower().replace(" ","")
        s = ''.join(filter(str.isalnum,s))
        n=len(s)-1
        for i in range(len(s)//2):
            if s[i]!=s[n]:
                return False
            n-=1
        return True
            

        
        
```
